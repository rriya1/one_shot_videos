```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline #to show the outputs below the commands and not as popups
```

```python
#plotting a line graph/chart
apples=[.234, 0.2525,0.366,0.745]
plt.plot(apples)
#x axis is the index and y is the value at that index
```

```python
plt.plot(apples); #placed semicolon to get rid of the unintended output
```

```python
years=[2010,2015,2016,2017]
plt.plot(apples,years);#graph got reversed, vlue f x axis and year is y axis
```

```python
plt.plot(years,apples); #original graph with x axis as years
```

```python
#putting labels for x axis and y axis
plt.xlabel('year')
plt.plot(years,apples);
```

```python
#plotting multiple lines whithin the same graph
oranges=[0.242,0.635,0.767,0.865]
#keeping the x axis same here
plt.plot(years,apples)
plt.plot(years,oranges)
plt.xlabel('year')
plt.ylabel('yeild per year')
#also including the legend to identify which line is which
plt.title('fruit yeild')
plt.legend(['apples','oranges'])
```

```python
#adding markers to the graph
plt.plot(years,apples, marker='o');
plt.plot(years,oranges);
plt.xlabel('year')
plt.ylabel('yeild per year')
plt.title('fruit yeild')
plt.legend(['apples','oranges'])
```

```python
#there are many options to style the graph like--> setting the color of the line
#line style, width of line, size of marker, color of marker(edge,face), opacity
plt.plot(years, apples, marker='s', c='b', ls='-', lw=2, ms=8, mew=2, mec='navy')
plt.plot(years, oranges, marker='o', c='r', ls='--', lw=3, ms=10, alpha=.5)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);

#there is also a shortcut for this, we genrally want to set the marker, line and color for the line so the shortcut is fmt string
#fmt=[marker][line][color]
grapes=[0.234,0.466,0.757,0.526]
plt.plot(years,grapes,'s--g');
```

```python
#if we do not specify the type of line in the shortcut then only the markers will be drawn
plt.plot(years,grapes,'sg');
plt.plot(years,apples,'or');
```

```python
#changing the size of the figure
plt.figure(figsize=(12,6))
plt.plot(years,grapes,'s--g');
plt.plot(years,apples,'.-r');
#improving the look of out graph by using seaborn library
#gets applied globally
#sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(12,6))
plt.plot(years,grapes,'s--g');
plt.plot(years,apples,'.-r');
#improving the look of out graph by using seaborn library
sns.set_style("darkgrid")
#we can change styling in marplotlib using rcparams
```

```python
#SCATTER PLOT #loading the default dataset available in sns
flowers_df=sns.load_dataset('iris') 
#it returns a pandas dataframe
flowers_df
```

```python
flowers_df.species.unique()
```

```python
plt.plot(flowers_df.sepal_length,flowers_df.sepal_width)
#there are about 150 values randomly in 4 units, so its very unclear
```

```python
#sns scatter plot
sns.scatterplot(data=flowers_df,x='sepal_length',y='sepal_width');
```

```python
#giving the hue argument
#s is for size of the points
sns.scatterplot(data=flowers_df,x='sepal_length',y='sepal_width',hue=flowers_df.species,s=90);
```

```python
#seaborn is built upon matplotlib so we can also call matplotlib functions over sns
#HISTOGRAM
#histogram is typically data fro a single column
plt.hist(flowers_df.sepal_width);
#divided into 10 intervals by default
```

```python
flowers_df.sepal_width.describe()
#visulaization of this with the histogram is way easier
```

```python
plt.hist(flowers_df.sepal_width,bins=15);
```

```python
import numpy as np
#specify the points at which bina should be created
plt.hist(flowers_df.sepal_width,bins=np.arange(3,5,0.25));
```

```python
plt.hist(flowers_df.sepal_width,bins=[1,5,8,]);
```

```python
#multiple histogram in one chart
#opacity matters here
#first one which will be called will be at bottom and then the next one on its top and the next one on top of thm both and so on
setosa_df = flowers_df[flowers_df.species == 'setosa']
versicolor_df = flowers_df[flowers_df.species == 'versicolor']
virginica_df = flowers_df[flowers_df.species == 'virginica']
```

```python
plt.hist(setosa_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
plt.hist(versicolor_df.sepal_width, alpha=0.3, bins=np.arange(2, 5, 0.25));
#in this the histograms are overlapping
```

```python
#stacking histograms on top of one another
plt.title('Distribution of Sepal Width')

plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], 
         bins=np.arange(2, 5, 0.25), 
         stacked=True);

plt.legend(['Setosa', 'Versicolor', 'Virginica']);
```
